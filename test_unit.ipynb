```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('test.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img = cv2.flip(img, 0)


plt.figure(figsize=(10,10))
plt.xlim(0, 350)
plt.ylim(0, 350)
plt.xticks(np.arange(0, 350, 50))
plt.yticks(np.arange(0, 350, 50))
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(linestyle='--')
plt.imshow(img)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(0)
data = np.random.normal(50, 10, 1000)

def z_score_normalization(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    return (data - mean) / std_dev

normalized_data = z_score_normalization(data)

def plot_normal_distribution(data, title):
    plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
    
    mu, std = norm.fit(data)

    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    plt.plot(x, p, 'k', linewidth=2)
    plt.title(title)
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.show()

plot_normal_distribution(data, "Original Data")
plot_normal_distribution(normalized_data, "Normalized Data (Z-score)")
```
