```python
#可以给一个jpg图片添加坐标系吗？

import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('test.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#将坐标系倒置
img = cv2.flip(img, 0)

#在坐标系中加入虚线背景
#坐标系横轴为x1，纵轴为x2
plt.figure(figsize=(10,10))
plt.xlim(0, 350)
plt.ylim(0, 350)
plt.xticks(np.arange(0, 350, 50))
plt.yticks(np.arange(0, 350, 50))
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(linestyle='--')
plt.imshow(img)
plt.show()
```
